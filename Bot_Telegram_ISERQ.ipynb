```python
import requests
import os

# Tus credenciales
TELEGRAM_BOT_TOKEN = ""
TELEGRAM_CHAT_ID = ""

# Mensaje de prueba
mensaje = "SIGUE FUNCANDO"

# URL de la API de Telegram
url = f"https://api.telegram.org/bot{TELEGRAM_BOT_TOKEN}/sendMessage"

# Payload con chat_id y texto
payload = {
    "chat_id": TELEGRAM_CHAT_ID,
    "text": mensaje,
    "parse_mode": "HTML"
}

# Enviar mensaje
response = requests.post(url, data=payload)

# Mostrar respuesta de Telegram
print(response.status_code)
print(response.json())
```
